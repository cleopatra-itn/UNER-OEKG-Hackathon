### Counting statistics

```python
diego = "/media/gaurish/angela/projects/UNER_HF/compare-entities/list_entities_EN_v1"
simone ="/media/gaurish/angela/projects/UNER_HF/compare-entities/same_as_wikipedia_en.tsv"
df_simone = pd.read_csv(simone,sep="\t")
df_diego = pd.read_csv(diego,sep="\t")
df_simone.columns= ["id","name"]
df_diego.columns= ["uner-name", "uner-class"]
df_diego["name"] =df_diego["uner-name"].map(str).apply(lambda x : "_".join(x.split()))
df_result = df_diego.join(df_simone.set_index('name'), on='name', how='inner')
```

### 1. Distribution of entities EN 

```python
df_result["uner-class"].value_counts(normalize=True).to_csv("uner-class-en-v1-distribution-percentage.tsv",sep="\t")
```

```python
df_result["uner-class"].value_counts().to_csv("uner-class-en-v1-distribution.tsv",sep="\t")
```

```python
pd.concat([df_result["uner-class"].value_counts(normalize=True),df_result["uner-class"].value_counts()],axis=1).to_csv("uner-class-en-v1-distribution-with-percentage.tsv",sep="\t")
```

```python
tmp = df_diego.join(df_simone.set_index('name'), on='name', how='outer')
tmp = tmp[tmp["id"].isnull()]
tmp = pd.concat([tmp["uner-class"].value_counts(normalize=True),
           tmp["uner-class"].value_counts(),
           df_diego["uner-class"].value_counts(),
           tmp["uner-class"].value_counts()/df_diego["uner-class"].value_counts()],axis=1)
tmp.columns= ["normalized-unmatched-count","unmatched-count","total-class-count","normalized-unmatched-count"]
tmp.to_csv("uner-class-en-v1-non-matched-distribution-with-percentage.tsv",sep="\t")
```
